```python
import os
import json
import pandas as pd

tweets = []

with open('CIPds.json', 'r') as input_file:
    for line in input_file:
        try:
            tweet = json.loads(line)
            tweets.append(tweet)
        except:
            continue
ids=[]
text=[]
created_at=[]
for tweet in tweets :
    if 'RT @BCCI' in tweet['text']:
        ids.append(tweet['id_str'])
        text.append(tweet['text'])
        text = [sub.replace('\n', '') for sub in text ]
        created_at.append(tweet['created_at'])

df=pd.DataFrame({'Ids':pd.Index(ids),
                 'Text':pd.Index(text),
                 'Created_at':pd.Index(created_at)})
#df.to_csv('BCCI.csv')
df
```
