```python
#Read original csv dataset
import pandas as pd  
import numpy as np
df = pd.read_csv("INDvsENGds.csv", engine = "python")
df
```

```python

#Create a seperate column for hashtags
import re, string, unicodedata
df['hashtag'] = df['Text'].apply(lambda x: re.findall(r'\B#\w*[a-zA-Z]+\w*', x))

#Remove mentions
df['Text_new'] = df['Text'].apply(lambda x: re.sub(r'@\w+', '', x))

#Remove &amp html entity
df['Text_new'] = df['Text'].apply(lambda x: re.sub(r'@\w+', '', x))

#Remove hashtag symbol
df['Text_new'] = df['Text_new'].apply(lambda x: re.sub('#', '', x))

#Remove emojis
df['Text_new'] = df['Text_new'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))

#Remove urls
df['Text_new'] = df['Text_new'].apply(lambda x: re.sub('https:\/\/.*', '',x))

from nltk.corpus import stopwords
stop = stopwords.words('english')

# Exclude stopwords 
df['Text_new'] = df['Text_new'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

df
```

```python
#Remove tweets that contains 4 words or less
i=0
index=[]
for text in df['Text_new'][0:21067]:
    number = len(text.strip().split(" ")) 
    if number < 5 :
        index.append(i)
    i+=1
```

```python
update_df = df.drop(index)
update_df.to_csv('INDvsENGds_updated.csv')
update_df
```
